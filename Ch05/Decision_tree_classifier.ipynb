# Decision Tree Implementation on Cancer Dataset (Breast Cancer)


Dataset source: sklearn

Features: Computed from a digitized image of a FNA of a breast mass. 
Target: Diagnosis (M = malignant, B = benign)

Class distribution: 357 benign, 212 malignant

Number of features: 30

# Load dataset

```python
from sklearn import datasets

#Load dataset
df = datasets.load_breast_cancer()
```

```python
df.data.shape
```

```python
# feature names
print("Features: ", df.feature_names)
```

```python
#label type of cancer('malignant' 'benign')
print("Labels: ", df.target_names)
```

# Creating training/testing datasets

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.3,random_state=109) 
print("shape of X_train=",X_train.shape)
print("shape of y_train=",y_train.shape)
print("shape of X_test=",X_test.shape)
print("shape of y_test=",y_test.shape)
```

# Decision Tree Classifier (gini)

## Model building

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
```

```python
classifier.fit(X_train, y_train)
```

## Predictions

```python
y_pred_gini = classifier.predict(X_test)
```

## Model evalution

```python
from sklearn.metrics import accuracy_score

print('Accuracy score (gini): {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

## Tree Visualization

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))

from sklearn import tree

tree.plot_tree(classifier.fit(X_train, y_train)) 
```

# Decision Tree Classifier (entropy)

## Model building

```python
classifier_entropy = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)

# fit the model
classifier_entropy.fit(X_train, y_train)
```

## Predictions

```python
y_entropy_pred = classifier_entropy.predict(X_test)
```

## Model evaluation

```python
from sklearn.metrics import accuracy_score

print('Accuracy score (entropy): {0:0.4f}'. format(accuracy_score(y_test, y_entropy_pred)))
```

## Tree visualization

```python
plt.figure(figsize=(15,8))

from sklearn import tree
tree.plot_tree(classifier_entropy.fit(X_train, y_train)) 
```

## Confusion matrix

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_entropy_pred)
print('Confusion matrix\n\n', conf_matrix)
```

## Classification report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_entropy_pred))
```
