# Support Vector Machine

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
# loading iris dataset
iris = datasets.load_iris()
```

```python
#feature selection
X = iris.data[:, :2]#feature
y = iris.target#target
```

```python
#mesh plot 
def plotSVC(title):
    # create a mesh to plot in
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    h = (x_max / x_min)/100
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.summer, alpha=0.8)#setting contours
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.rainbow)#plotting scatter plot
    plt.xlabel('Sepal length')#x-axis label
    plt.ylabel('Sepal width')#y-axis label
    plt.xlim(xx.min(), xx.max())#setting limiting bounds
    plt.title(title) #title 
    plt.show()
```

# Kernel Types

## Linear Kernel



```python
# Support vector classifier with linear kernel
svc = svm.SVC(kernel='linear')
svc.fit(X, y)
plotSVC('kernel=' + str('linear'))
```

## Polynomial Kernel


```python
# support vector classifier with polynomial kernel
svc = svm.SVC(kernel='poly')
svc.fit(X, y)
plotSVC('kernel=' + str('poly'))
```

## Radial Basis Function Kernel



```python
#support vector with rbf kernel
svc = svm.SVC(kernel='rbf')
svc.fit(X, y)
plotSVC('kernel=' + str('rbf'))
```

## Sigmoid kernel Function


```python
#support vector classifier with sigmoid kernel
svc = svm.SVC(kernel='sigmoid')
svc.fit(X, y)
plotSVC('kernel=' + str('sigmoid'))
```

We can see that the sigmoid kernel fails to distinguish for the multi-class problem.

# Hyperparameters tuning

## C : float, default=1.0



```python
C_values = [0.1, 1, 10, 100, 1000, 10000]#different C values for model training
for ci in C_values:#looping through all C values and plotting
    svc = svm.SVC(kernel='rbf', C=ci)
    svc.fit(X, y)
    plotSVC('C = ' + str(ci))
```

## degree: int, default=3


```python
D = [0, 1, 2, 3, 4, 5]#setting different Degree values for polynomial kernel
for degree in D:#looping each degree value from the list and visualization
    svc = svm.SVC(kernel='poly', degree=degree)
    svc.fit(X, y)
    plotSVC('degree = ' + str(degree))
```

We get no hyperplane if we set degree =0.

## gamma: default = 'scale'

```python
G = ['auto',0.1, 1, 10, 100]#setting different Gamma values for RBF kernel
for gamma in G:#looping each value from the list 
    svc = svm.SVC(kernel='rbf', gamma=gamma)
    svc.fit(X, y)
    plotSVC('gamma = ' + str(gamma))
```

```python
svc = svm.SVC(kernel='poly', gamma='scale')
svc.fit(X, y)
plotSVC('gamma = ' + 'scale')
```

```python
svc = svm.SVC(kernel='poly', gamma='auto')
svc.fit(X, y)
plotSVC('gamma = ' + 'auto')
```

## decision_function_shape: default = 'ovr;

```python
svc = svm.SVC(kernel='poly', decision_function_shape='ovo')
svc.fit(X, y)
plotSVC('decision_function_shape = ' + 'ovo')
```

```python
svc = svm.SVC(kernel='poly', decision_function_shape='ovr')
svc.fit(X, y)
plotSVC('decision_function_shape = ' + 'ovr')
```
