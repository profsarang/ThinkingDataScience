```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns;
from sklearn.metrics import pairwise_distances
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids

sns.color_palette("Set2")
%matplotlib inline
```

## ***K-Means Clustering***

```python
def plotdata(X):
  plt.rcParams["figure.figsize"] = (10,6)   
  plt.grid()
  plt.xlim(-2, 10)
  plt.scatter(X[:, 0], X[:, 1], s=100, c='blue');
```

```python
def plotdata_with_centroids(X, centers):
  plt.rcParams["figure.figsize"] = (10,6)   
  plt.grid()
  plt.xlim(-2, 10)
  print(centers)
  plt.scatter(X[:, 0], X[:, 1], s=100, c='blue');
  plt.scatter(centers[:, 0], centers[:, 1], marker='X',s=500, c='red');
```

## Sample dataset

```python
X, Y = make_blobs(n_samples=30, random_state=0, centers=1)
```

```python
X.shape
```

```python
plotdata(X)
```

## Dataset after adding outliers

```python
add_outliers = np.array([[6, 7], [6, 5], [8, 9], [9, 6], [8, 7], [9, 8], [7, 7], [8, 5], [8, 9], [7, 8], [8, 7], [9, 8]])
```

```python
X1 = np.concatenate((add_outliers , X))
```

```python
X1.shape
```

```python
plotdata(X1)
```

## K-Means on original dataset

```python
kmeans = KMeans(n_clusters=1)
kmeans.fit(X)
km_pred = kmeans.predict(X)
centers = kmeans.cluster_centers_
```

```python
plotdata_with_centroids(X, centers)
```

## K-Means with outliers

```python
kmeans = KMeans(n_clusters=1)
kmeans.fit(X1)
km_pred = kmeans.predict(X1)
centers = kmeans.cluster_centers_
```

```python
plotdata_with_centroids(X1, centers)
```

## ***K-Medoids Clustering***

```python
model = KMedoids(metric="euclidean", n_clusters=1)
model.fit(X)
kmed_pred = model.predict(X)
centers = model.cluster_centers_
```

```python
plotdata_with_centroids(X, centers)
```

```python
model = KMedoids(metric="euclidean", n_clusters=1)
model.fit(X1)
kmed_pred = model.predict(X1)
centers = model.cluster_centers_
```

```python
plotdata_with_centroids(X1, centers)
```
