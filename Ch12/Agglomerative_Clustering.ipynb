# Agglomerative Clustering

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

## Annual income/Spending score Clustering







```python
X = dataset.iloc[:, [3,4]].values
```

### Two clusters

```python
# creating object hc of class AgglomerativeClustering()
hc=AgglomerativeClustering(n_clusters=2, affinity="euclidean",linkage="ward")
```

```python
# it gives an array which tells as to which data point belongs to which cluster
y_hc=hc.fit_predict(X)
```

```python
# converting x into ndarray
x=np.array(X)
```

```python
# Visualising the clusters
fig = plt.figure(figsize=(10, 6))
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('Customers - two clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

### Three clusters

```python
hc=AgglomerativeClustering(n_clusters=3, affinity="euclidean",linkage="ward")
y_hc=hc.fit_predict(X)
x=np.array(X)
```

```python
# Visualising the clusters
fig = plt.figure(figsize=(10, 6))
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Customers - three clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

### Four clusters

```python
hc=AgglomerativeClustering(n_clusters=4, affinity="euclidean",linkage="ward")
y_hc=hc.fit_predict(X)
x=np.array(X)
```

```python
# Visualising the clusters
fig = plt.figure(figsize=(10, 6))
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 100, c = 'black', label = 'Cluster 4')
plt.title('Customers - four clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

# Age/Spending score clustering

```python
Y = dataset.iloc[:, [2,4]].values
```

```python
hc=AgglomerativeClustering(n_clusters=3, affinity="euclidean",linkage="ward")
yy_hc=hc.fit_predict(Y)
```

```python
y=np.array(Y)
```

```python
# Visualising the clusters
fig = plt.figure(figsize=(10, 6))
plt.scatter(y[yy_hc == 0, 0], y[yy_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(y[yy_hc == 1, 0], y[yy_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(y[yy_hc == 2, 0], y[yy_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')

plt.title('Clusters of customers')
plt.xlabel('Customer Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

# Dendogram

```python
# Visualising the dendrogram
fig = plt.figure(figsize=(10, 6))
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Eucledian distance")
plt.show()
```
