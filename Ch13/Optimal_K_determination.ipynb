```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.mixture import GaussianMixture
import pandas as pd
import seaborn as sns
import numpy as np
```

# 3-Blob dataset

```python
# Generating 5000 samples using make_blobs
X, Y = make_blobs(n_samples = 1000, n_features = 3, cluster_std = 0.9, random_state = 0)
```

```python
plt.rcParams["figure.figsize"] = (6,6)   
plt.grid()
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis');
```

## Compute bic 

```python
bic_array = []

K=[2,3,4,5]
cv_types = ['spherical', 'tied', 'diag', 'full']
for cov_type in cv_types:
    for k in K:
        gmm = GaussianMixture(n_components =k, covariance_type=cov_type)
        gmm.fit(X)
        bic = gmm.bic(X)
        bic_array.append(bic)
```

## Extract data

```python
width = 0.2
mk = np.array(K)
x = np.array(bic_array)
spherical_data = x[0:4]
tied_data = x[4:8]
diag_data = x[8:12]
full_data = x[12:16]
```

## Plot bic v/s K

```python
plt.figure(figsize=(12,6))
plt.bar(mk-width, spherical_data, width, label='spherical')
plt.bar(mk, tied_data, width, label='tied')
plt.bar(mk+width, diag_data, width, label='diag')
plt.bar(mk+width*2, full_data, width, label='full')
plt.xticks(mk)
plt.xlabel("K")
plt.ylabel("Bayesian Information")
plt.title ("Comparison for different covariance")
plt.legend()
plt.show()
```

## Compute bic/aic

```python
bic_array = []
aic_array = []

for k in K:
    gmm = GaussianMixture(n_components=k)
    gmm.fit(X)
    bic_array.append(gmm.bic(X))
    aic_array.append(gmm.aic(X))
```

## Plot bic/aic v/s K

```python
plt.figure()
plt.plot(K, bic_array, label='bic')
plt.plot(K, aic_array, label='aic')
plt.legend()
plt.xlabel('K')
plt.ylabel('AIC/BIC')
plt.title ("AIC/BIC versus K")
```
