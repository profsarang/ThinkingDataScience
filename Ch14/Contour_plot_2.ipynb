# Contours for different bandwidths

```python
from sklearn.datasets import make_blobs

from pandas import DataFrame
import seaborn as sns
```

## Generating dataset

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2)
```

```python
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
```

```python
df.head()
```

## Dataset plot

```python
sns.scatterplot(x="x", y="y", data=df)
```

## Bandwidth = 1.0 (default)

```python
sns.kdeplot(data=df, x="x", y="y")
```

## Bandwidth = 2.0

```python
sns.kdeplot(data=df, x="x", y="y",bw_adjust = 2)
```

## Bandwidth = 0.8

```python
sns.kdeplot(data=df, x="x", y="y",bw_adjust=0.8)
```
