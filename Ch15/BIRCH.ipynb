# Birch

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch
import pandas as pd
import seaborn as sns
from numpy import unique
```

## Data genertion

```python
# Generating 5000 samples using make_blobs
X, Y = make_blobs(n_samples = 5000, n_features = 3, cluster_std = 5, random_state = 128)
```

```python
plt.scatter(X[:,0], X[:,1],
            cmap='red',
            alpha = 0.7,
            edgecolors = 'b'
            )

plt.grid()
plt.show()
```

## Default parameters

```python
# Creating the BIRCH clustering model
model = Birch().fit(X) 
```

```python
pred = model.predict(X)  
```

```python
pred
```

```python
clusters = unique(pred)
```

```python
def plottingFun():
  print("Number of clusters: ", len(clusters))
  plt.scatter(X[:, 0], X[:, 1], c=pred, cmap='viridis')
  plt.grid()
```

```python
plottingFun()
```

## Effect of n_clusters

```python
model = Birch(n_clusters = 4) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = 5) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = 6) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = None) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

## Effect of branching factor

```python
model = Birch(n_clusters = 6, branching_factor = 50) #default 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = 6, branching_factor = 10) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = 6, branching_factor = 100) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = 6, branching_factor = 200) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

## Effect of threshold

```python
model = Birch(n_clusters = 6, threshold = 0.5) # default
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = 6, threshold = 0.1) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = 6, threshold = 1.0) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
model = Birch(n_clusters = 6, threshold = 2.0) 
model.fit(X)  
pred = model.predict(X)  
clusters = unique(pred)
plottingFun()
```

```python
data = pd.read_csv("/content/Mall_Customers.csv")
```

```python
data["Income"] = data[["Annual Income (k$)"]]
data["Spending"] = data[["Spending Score (1-100)"]]
data = data[["Income", "Spending"]]
```

```python
model = Birch(branching_factor=30, n_clusters=5, threshold=2.5)
model.fit(data)
pred = model.predict(data)
plt.scatter(data["Income"], data["Spending"], c=pred, cmap='rainbow', alpha=1.0, edgecolors='b')
plt.title('Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
