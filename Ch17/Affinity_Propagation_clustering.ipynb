```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AffinityPropagation
from sklearn.datasets import make_blobs
from sklearn import metrics
import matplotlib.pyplot as plt
from itertools import cycle
```

# Blob dataset

```python
centers = [[1,1],[-1,-1],[1,-1]]
X, Y = make_blobs(
    n_samples=300, centers=centers, cluster_std=0.5, random_state=0
)
```

```python
plt.rcParams["figure.figsize"] = (6, 6)   
plt.grid()
plt.title ('Data distribution')
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis', c='blue');
```

## Clustering

```python
model = AffinityPropagation(preference=-50).fit(X)
```

```python
center_indices = model.cluster_centers_indices_
number_of_clusters = len(center_indices)
print("Estimated number of clusters: %d" % number_of_clusters)

number_of_clusters = len(center_indices)
```

```python
labels = model.labels_
print("Homogeneity: %0.3f" % metrics.homogeneity_score(Y, labels))
```

```python
labels
```

```python
def plot_clusters():  
  
  colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
  
  for k, col in zip(range(number_of_clusters), colors):
      class_members = labels == k
      cluster_center = X[center_indices[k]]
      plt.plot(X[class_members, 0], X[class_members, 1], col + '.')
      plt.plot(
        cluster_center[0], 
        cluster_center[1], 
        'o',
        markerfacecolor = col, 
        markeredgecolor ='k',
        markersize = 10
      )
  
      for x in X[class_members]:

        plt.plot(
            [cluster_center[0], x[0]], 
            [cluster_center[1], x[1]], 
            col)
  
  plt.title('Estimated number of clusters: % d' % number_of_clusters)
  plt.show()
```

```python
plot_clusters()
```

## preference1 (-40)

```python
model = AffinityPropagation(preference=-40).fit(X)
center_indices = model.cluster_centers_indices_
number_of_clusters = len(center_indices)
labels = model.labels_
plot_clusters()
```

## Annual Income / Spending score clustering

```python
X = dataset.iloc[:, [2, 3]].values
```

```python
ap = AffinityPropagation()
ap_model = ap.fit(X)
ap_y_pred = ap.predict(X)
plt.figure(figsize=(6, 6))
plt.title('Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.scatter(X[:,0], X[:,1],c=ap_y_pred, cmap='ocean')
plt.title("Affinity Propagation Model")
```
