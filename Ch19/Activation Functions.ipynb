#Activation Functions

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#Linear (Identity)
Equation:
f(x) = x

Derivative: f’(x) = 1

Range: (-∞, +∞)

```python
x = np.linspace(-10, 10)
y = x
plt.plot(x, y)
plt.axis('tight')
plt.title('Activation Function :Linear')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#Rectified Linear Unit (ReLU)

Equation: f(x) = a =max(0,x)

Derivative: f’(x) = { 1 ; if z>0, 
                     ow  0 ; if z<0 



Range: (0, +∞)

```python
def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.title('Activation Function :RELU')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#Lecky Relu

Equation: f(x)= a = max(0.01x, x)

Derivative: f’(x) = {0.01 ; if z<0, 1 ; otherwise}

Range: (0.01, +∞)

```python
def Lecky_relu(x,a=0.01):
  x2=[]
  for i in x:
    if i<0:
      i=a*i
    x2.append(i)
  return x2    
```

```python
x = np.linspace(-10,10)
plt.plot(x, Lecky_relu(x))
plt.title('Activation Function :Lecky Relu')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#Parametric Rectified Linear Unit (PReLU)

Equation: f(x) = x,  if x>=0,  else ax, x<0

Derivative: f'(x) = 1 if x>=0, else  a, x<0


```python
def Param_relu(x,a=0.1):
  x2=[]
  for i in x:
    if i<0:
      i=a*i
    x2.append(i)
  return x2    
```

```python
x = np.linspace(-10, 10)
y= Param_relu(x,a=0.1)
plt.plot(x, y)
plt.title('Activation Function :Parametric ReLU')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#Exponential Linear Unit

Equation: f(X) = a(exp(x)-1) if x<0,  else x if x>= 0

Derivative: f'(X) = f(x)+a if x<0,  else 1 if x>=0

```python
def ELU(x,a=1.0):
  x3=[]
  for i in x:
    if i<0:
      i=a*(np.exp(i)-1)
    x3.append(i)
  return x3   
```

```python
x = np.linspace(-10, 10)
y= ELU(x,a=1.0)
plt.plot(x, y)
plt.title('Activation Function :Parametric ReLU')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#Softmax
Equation: f(x) = S= eˣᵢ / (Σⱼ₌₀ eˣᵢ)

Range: (0, 1)

Derivitive:  DjSi = Si(δij – Sj)

Where, δij  = 1 if i= j,   else 0 if i≠j

(https://eli.thegreenplace.net/2016/the-softmax-function-and-its-derivative/)

```python
def softmax(x):
    """ applies softmax to an input x"""
    e_x = np.exp(x)
    return e_x / e_x.sum()
```

```python
x = np.linspace(-10, 10)
y = softmax(x)
plt.plot(x,y)
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```
