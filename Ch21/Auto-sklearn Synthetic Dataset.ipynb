```python
from autosklearn.classification import AutoSklearnClassifier
```

```python
from sklearn.model_selection import train_test_split
```

## Dataset for binary classification

```python
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=500, n_features=10, n_informative=5, n_classes=2, n_redundant=5, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

model = AutoSklearnClassifier(time_left_for_this_task=4*60, per_run_time_limit=30, n_jobs=-1)
model.fit(X_train, y_train)
print(model.sprint_statistics())

y_hat = model.predict(X_test)
acc = accuracy_score(y_test, y_hat)
print("Accuracy: %.3f" % acc)
```

## Dataset for multi-class classification

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=10, n_informative=5, n_classes=4, n_redundant=5, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

model = AutoSklearnClassifier(time_left_for_this_task=4*60, per_run_time_limit=30, n_jobs=-1)
model.fit(X_train, y_train)
print(model.sprint_statistics())

y_hat = model.predict(X_test)
acc = accuracy_score(y_test, y_hat)
print("Accuracy: %.3f" % acc)
```

The accuracy obtained is very low. At the time of this writing, the library does not fully support the multi-class classification.
